```python
import torch
from diffusers import StableDiffusionInpaintPipeline
from PIL import Image
import numpy as np

# Load the model
model_id = "stabilityai/stable-diffusion-2-inpainting"
pipe = StableDiffusionInpaintPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to("cuda")

# Load the image
image_path = "/content/assignment.jpg"
image = Image.open(image_path)
image = image.resize((1024, 1024))

# Define the outpainting function
def outpaint_image(image, pipe, padding=128):
    width, height = image.size
    new_width, new_height = width + 2 * padding, height + 2 * padding
    new_image = Image.new("RGB", (new_width, new_height), (255, 255, 255))
    new_image.paste(image, (padding, padding))

    # Convert new_image to tensor and create mask
    image_np = np.array(new_image)
    mask = np.ones((new_height, new_width), dtype=np.float32)
    mask[padding:padding+height, padding:padding+width] = 0

    # Run the inpainting pipeline
    result = pipe(prompt="", image=new_image, mask_image=Image.fromarray((mask * 255).astype(np.uint8)), guidance_scale=7.5, num_inference_steps=50).images[0]
    result = result.resize((1280, 1280))

    return result

# Perform outpainting
outpainted_image = outpaint_image(image, pipe, padding=128)

# Save the outpainted image
outpainted_image.save("/content/outpainted_image3.png")
```
